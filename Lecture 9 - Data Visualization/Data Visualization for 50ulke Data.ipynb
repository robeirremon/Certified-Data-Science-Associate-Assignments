# Data Visualization for 50ulke Dataset

```python
import seaborn as sns
import pandas as pd
```

```python
%matplotlib inline
sns.set(rc={'figure.figsize':[7,7]},font_scale=1.2)
```

```python
ulke = pd.read_csv('50ulke.csv',index_col='Countries')
ulke.drop('Unnamed: 0',axis=1,inplace=True)
ulke.head()
```

```python
ulke.tail()
```

```python
ulke.info()
# numerial(National Income , Growth , Per person)
# categorical(Countries,Continent)
```

```python
ulke.describe()
```

## Data Visualization for Numerial Columns
* 'National Income'
* 'Growth'
* 'Per person'

```python
sns.distplot(ulke['National Income'],bins=50)
#National Income is very low in most of the countries 
```

```python
sns.kdeplot(ulke['National Income'],shade=True)
#std for data is at 3448.554429
```

```python
sns.distplot(ulke['Growth'],bins=50)
#Growth is the most approx in the range between (1.5,3), std at 2
```

```python
sns.kdeplot(ulke['Growth'],shade=True)
#std for data is at 2.053468
```

```python
sns.distplot(ulke['Per person'],bins=50)
#most of the income per person for most countries is very low 
```

```python
sns.kdeplot(ulke['Per person'],shade=True)
#std for data is at 25094.131496
```

```python
sns.jointplot(x='Growth',y='National Income',data=ulke,kind='scatter')
# they are independant as Growth inc National Income doesn't affect we will check with correlation using heatmat later
#there is outlier in National Income it's United States then China lower than it.
```

```python
sns.jointplot(x='Growth',y='Per person',data=ulke,kind='scatter')
#most of the point at lower part of income Per person (0,20000) and in growth in range (1.5,3)
#there is an outlier where both features are very high
#there are two countries where they income are very very high but the growth is below avg so they are good countries to live in.
```

```python
sns.jointplot(x='Per person',y='National Income',data=ulke,kind='scatter')
#Obviously there are two outliers of US & China of V.High National Income 
#but the income Per person is only high in US but low in China.
'''
There are three countries where the National income is very low so the country is poor
but the  strange thing is the income per person is still high 
so people there are not affected by the National situations
'''
```

```python
sns.pairplot(ulke,vars=['Growth','Per person'],height=6,aspect=1.2,hue='Continent')
```

```python
sns.pairplot(ulke,vars=['National Income','Per person'],height=6,aspect=1.2,hue='Continent')
```
