```python
import seaborn as sns

%matplotlib inline
sns.set(rc={'figure.figsize':[9,9]},font_scale=1.2)
```

```python
titanic= sns.load_dataset('titanic')
```

```python
titanic.head(10)
```

```python
titanic.tail(10)
```

```python
titanic.info()
```

### Numerical Plot

```python
#numerical data : age,fare  
sns.distplot(titanic.age,bins=30)
sns.kdeplot(titanic.age,shade=True)
# mean of age 29.699118 approx 30 years
titanic.age.value_counts().head(15)
```

```python
sns.distplot(titanic.fare)
sns.kdeplot(titanic.age,shade=True)
titanic.fare.value_counts().head(15)
# mean of fare is 32.204208
```

```python
sns.jointplot(x='age',y='fare',data=titanic,height=10)
```

```python
sns.pairplot(titanic,vars=['age','fare'],aspect=1.4,height=9)
```

### Categorical Plot

data : sex , who , alive , alone , embark_town

```python
sns.boxplot(x='sex',y='age',data=titanic)
```

```python
sns.boxplot(x='alive',y='age',data=titanic)
```

```python
sns.boxplot(x='embark_town',y='age',data=titanic)
```
