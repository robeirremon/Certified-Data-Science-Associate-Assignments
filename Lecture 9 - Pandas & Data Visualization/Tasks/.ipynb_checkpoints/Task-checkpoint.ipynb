# csv file 

```python
df2=pd.read_csv('Ecommerce Purchases.csv')
```

```python
df2.info()
```

```python
df2['AM or PM'].value_counts()
```

```python
df2['Purchase Price'].describe()
```

```python
def extract_mail(text):
    _,txt=text.split('@')
    return txt
```

```python
df2['mail_vendor'] = df2['Email'].apply(extract_mail)
```

```python
df2['mail_vendor'].value_counts()
```

## Data Visualization (Seaborn)

```python
import seaborn as sns
```

```python
df2.info()
```

```python
df2['X'] = df2['Lot'].apply(lambda x : int(x.split()[0]))
```

```python
df2.info()
```

```python
sns.distplot(df2['X'],kde=False,color='m')
```

```python
sns.distplot(df2['Purchase Price'])
```

```python
sns.kdeplot(df2['X'], shade=True, color='m')
```

```python
sns.jointplot(x='X', y='Purchase Price', data=df2, kind='scatter')
```

```python
sns.jointplot(x='X', y='Purchase Price', data=df2, kind='kde')
```

```python
sns.pairplot(df2)
```
