# XO Game 

### Mode 1 Human to Human

```python
def draw_board(b):
    board_shape = '''
     _______ _______ _______ 
    |       |       |       |
    |   {}   |   {}   |   {}   |
    |_______|_______|_______|
    |       |       |       |
    |   {}   |   {}   |   {}   |
    |_______|_______|_______|
    |       |       |       |
    |   {}   |   {}   |   {}   |
    |_______|_______|_______|
    
    '''.format(b[0],b[1],b[2],b[3],b[4],b[5],b[6],b[7],b[8])
    print(board_shape)
```

```python
# Function to take the input and check that the number is in range from 0 to 8
def input_check(val):
    while True :
        num = input(f"Player {val} : Enter Your Position \n")
        try :
            num = int(num)
            if num in range(9):
                return num
            else:
                print("Please Enter a valid Number\n")
                continue
        except :
            print("That isn't a number\n")
            continue
```

```python
def player_X(board_num):
    x = input_check("X")
    if board_num[x] =='X' or board_num[x] == 'O':
        print("This number is already chosen")
        player_X(board_num)
    else:
        board_num[x] = "X"
        draw_board(board_num)
    
    return board_num
```

```python
def player_O(board_num):
    o = input_check("O")
    if board_num[o] =='X' or board_num[o] == 'O':
        print("This number is already chosen")
        player_O(board_num)
    else:
        board_num[o] = "O"
        draw_board(board_num)
    
    return board_num
```

```python
def check_winner(board_num,counter = 0):
    win_num = ((0,1,2),(3,4,5),(6,7,8),(0,3,6),(1,4,7),(2,5,8),(0,4,8),(2,4,6))
    
    for i in win_num :
        if board_num[i[0]]==board_num[i[1]]==board_num[i[2]] == 'X':
            print("Player X Win")
            print('Congratulations!')
            return True
        elif board_num[i[0]]==board_num[i[1]]==board_num[i[2]] == 'O':
            print("Player O Win")
            print('Congratulations!')
            return True
    
    return False
```

```python
def game():
    board_num = [i for i in range(9)]
    draw_board(board_num)
    flag = False
    counter = 0
    
    while not flag:
        board_num = player_X(board_num)
        flag = check_winner(board_num)
        if flag == True :
            break

        counter +=1
        if counter == 5:
            print("Draw Game")
            break
    
        board_num = player_O(board_num)
        flag = check_winner(board_num)
        if flag == True :
            break
    
```

```python
game()
```

### Mode 2 Human to Computer

```python
import random

def computer(board_num,value):
    num = random.choice(board_num)
    if num =='X' or num == 'O':
        computer(board_num,value)
    else:
        board_num[num] = value
    return board_num
```

```python
def game_2():
    board_num = [i for i in range(9)]
    draw_board(board_num)
    flag = False
    counter = 0
    
    while not flag:
        board_num = player_X(board_num)
        flag = check_winner(board_num)
        if flag == True :
            break

        counter +=1
        if counter == 5:
            print("Draw Game")
            break
    
        board_num = computer(board_num,"O")
        draw_board(board_num)
        flag = check_winner(board_num)
        if flag == True :
            break
    
```

```python
game_2()
```

### Mode 3 Human to Intelligent Computer

```python
def check_lose_win(board_num,val,player):
    #value is 'X' or 'O' 
    #use 'X' in case of check loosing
    #use 'O' in case of check winning
    flag = False
    win_num = ((0,1,2),(3,4,5),(6,7,8),(0,3,6),(1,4,7),(2,5,8),(0,4,8),(2,4,6))
    for i in win_num :
        if board_num[i[0]] == board_num[i[1]]==val and board_num[i[2]]==i[2] : # board_num[i[2]]!='X' and board_num[i[2]]!='O': 
            board_num[i[2]]= player
            flag = True
            break
        elif board_num[i[1]] == board_num[i[2]]==val and board_num[i[0]]==i[0]:
            board_num[i[0]]= player
            flag = True
            break
        elif board_num[i[0]] == board_num[i[2]]==val and board_num[i[1]]==i[1]:
            board_num[i[1]]= player
            flag = True
            break
    return board_num,flag
```

```python
import random

def intel_computer(board_num):
    board_num,flag = check_lose_win(board_num,'O','O') # Check on winning first
    if flag == False :
        board_num,flag = check_lose_win(board_num,'X','O') # Then check on loosing
    if flag == False:
        better_choice = [0,2,4,6,8]
        num = random.choice(better_choice)
        if board_num[num] == num : 
            board_num[num]='O'
        else :
            board_num = computer(board_num,"O")
    print("Player O")
    draw_board(board_num)
    return board_num
```

```python
def game_3():
    board_num = [i for i in range(9)]
    draw_board(board_num)
    flag = False
    counter = 0
    
    while not flag:
        board_num = player_X(board_num)
        flag = check_winner(board_num)
        if flag == True :
            break

        counter +=1
        if counter == 5:
            print("Draw Game")
            break
    
        board_num = intel_computer(board_num)
        flag = check_winner(board_num)
        if flag == True :
            break
    
```

```python
game_3()
```
